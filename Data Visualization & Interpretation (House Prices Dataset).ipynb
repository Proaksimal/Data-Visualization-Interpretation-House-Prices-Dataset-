# 1.&nbsp; Count of one categorical feature

```python
roof_style = df['RoofStyle'].value_counts().sort_index()
roof_style
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(roof_style.index, roof_style.values, color='green')

ax.set_title("Roof Style Distribution", fontsize=16, fontweight='bold')
ax.set_xlabel("Style", fontsize=12)
ax.set_ylabel("Quantity", fontsize=12)

ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Dari distribusi data Roof Style di atas, dapat kita berikan beberapa kesimpulan berikut:


1.   Jenis Atap Gable merupakan roof style yang paling umum digunakan pada dataset, sebanyak 1141 unit. Atap Gable atau dalam bahasa Indonesia disebut Atap Pelana, merupakan jenis struktur atap berbentuk dua sisi miring menyerupai segitiga. Pada dataset, Atap Gable memiliki jumlah terbanyak, bisa jadi dikarenakan kepraktisannya dalam dibuat dan biayanya yang paling murah karena simpel dan hanya memiliki 2 sisi.
2. Jenis Atap Hip merupakan roof style kedua paling umum di bawah Atap Gable, dengan total sebanyak 286 unit. Atap Hip atau dalam bahasa Indonesia disebut Atap Limas adalah bentuk yang lebih sempurna dibanding Atap Gable. Jenis atap ini biasa dipilih karena tampilannya yang estetis. Namun dari segi harga, lebih mahal dibanding Atap Gable.  
3. Jenis Atap Flat (13 Unit), Gambrel (11 Unit), Mansard (7 Unit), dan Shed (2 Unit) memiliki jumlah yang jauh lebih sedikit dibanding Atap Gable (1141 Unit) dan Atap Hip (286 Unit). Beberapa alasannya bisa jadi dikarenakan modelnya yang tidak lumrah, kurang estetik, ataupun biaya proses pembuatannya yang lebih tinggi.

Dari hasil kesimpulan di atas, dapat disimpulkan jika mayoritas perumahan menggunakan Atap Gable dan Atap Hip. Sehingga kita dapat mengetahui preferensi pelanggan terhadap jenis atap dalam mencari rumah hunian.


# 2.&nbsp; Distribution of one continuous feature

```python
garage_area_count = df['GarageArea'].value_counts().sort_index()
garage_area_count
```

```python
df['GarageArea'].value_counts()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['GarageArea'], bins=100, color='red', kde=True)

plt.title("Garage Area", fontsize=16, fontweight='bold')
plt.xlabel("Area", fontsize=12)
plt.ylabel("Quantity", fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Dari distribusi Garage Area di atas, dapat kita berikan analisis sebagai berikut:



1.   Terdapat 81 data dari rumah yang memiliki luas garasi bernilai 0. Hal ini dapat menunjukkan jika rumah tersebut tidak memiliki garasi. 81 data ini juga merupakan yang paling tinggi dibandingkan nilai yang lain, sehingga dapat diartikan bahwa cukup banyak rumah yang ada pada dataset tidak memiliki garasi sama sekali.
2.   Pada luas garasi dengan nilai kecil sampai menengah memiliki jumlah yang cukup tinggi. Hal ini menunjukkan bahwa sebagian besar rumah pada dataset memiliki garasi yang berukuran standar.
3. Terdapat data-data di sisi kanan (membentuk ekor kanan) pada visualisasi distribusi Garage Area. Hal ini menunjukkan adanya sebagian kecil rumah yang memiliki luas garasi yang lebih besar dari rata-rata.

# 4.&nbsp; Continuous-continuous

```python
df['TotalBsmtSF_interval'] = pd.cut(df['TotalBsmtSF'], bins=15)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))

sns.scatterplot(x='TotalBsmtSF', y='SalePrice', data=df, hue='TotalBsmtSF_interval', palette='Set1')
plt.title('Distribution of Basement Area by Sale Price', fontsize=16)
plt.xlabel('Square Feet of Basement', fontsize=12)
plt.ylabel('Sale Price', fontsize=12)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

Dari distribusi data hubungan antar luas basement (TotalBsmtSF) dan harga jual (SalePrice), dapat dibuat beberapa analisis sebagai berikut:



1.   Sekilas dapat dilihat bahwa semakin luas basement dari suatu bangunan, maka harganya juga cenderung semakin tinggi. Maka antara faktor luas basement dengan harga jual memiliki hubungan korelasi yang positif. Karena semakin meningkat variabel luas basement, maka semakin meningkat juga variabel harga jual
2.   Ada beberapa bangunan yang tidak memiliki basement (luas basement = 0), dan harga jual yang bervariasi juga. Hal ini menunjukkan bahwa basement bukan satu-satunya faktor yang memengaruhi harga jual suatu bangunan, tetapi ada faktor lain juga (contohnya bisa seperti: luas garasi, tipe bangunan, dan jenis atap seperti poin-poin soal sebelumnya)
3. Sekilas, dapat dilihat jika tipe harga dari suatu bangunan terbanyak berada pada kisaran rentang 100.000 sampai 250.000. Dan luas basement terbanyak berada pada kisaran rentang 500 sampai 1800. Maka dari data tersebut didapat quick insight bahwa kebanyakan bangunan memiliki range tersebut.
